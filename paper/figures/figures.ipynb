```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Mostly flat baseline with small noise
n = 1000
x = np.arange(n)
y = 0.01 * np.random.randn(n)

# Add one or two big spikes
y[250:350] += 0.2
y[700:800] += 0.1

fig, axes = plt.subplots(1, 2, figsize=(16, 4), sharey=True)

for ax in axes:
    ax.plot(x, y)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_title("")
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

axes[0].axvspan(120, 240, color = 'green', alpha = 0.2)
axes[0].axvspan(240, 300, color = 'purple', alpha = 0.2)
axes[1].axvspan(650, 700, color = 'green', alpha = 0.2)
axes[1].axvspan(700, 800, color = 'purple', alpha = 0.2)
axes[1].axvspan(800, 850, color = 'green', alpha = 0.2)

plt.tight_layout()
plt.show()
```

```python
# Mostly flat baseline with small noise
n = 800
x = np.arange(n)
y = 0.01 * np.random.randn(n)

# Add one or two big spikes
y[50:150] += 0.2
y[250:350] += 0.2
y[450:550] += 0.2
y[650:750] += 0.2

fig, axes = plt.subplots(1, 2, figsize=(16, 4), sharey=True)

axes[0].plot(x, y)
axes[1].plot(x[0:200], y[400:600], c = '#1f77b4')
axes[1].plot(x[200:400], y[600:800], c = '#1f77b4')
axes[1].plot(x[400:600], y[0:200], c = '#1f77b4')
axes[1].plot(x[600:800], y[200:400], c = '#1f77b4')

for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_title("")
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

axes[0].axvspan(0, 200, color = 'green', alpha = 0.2)
axes[0].axvspan(200, 400, color = 'red', alpha = 0.2)
axes[0].axvspan(400, 600, color = 'purple', alpha = 0.2)
axes[0].axvspan(600, 800, color = 'yellow', alpha = 0.2)
axes[1].axvspan(0, 200, color = 'purple', alpha = 0.2)
axes[1].axvspan(200, 400, color = 'yellow', alpha = 0.2)
axes[1].axvspan(400, 600, color = 'green', alpha = 0.2)
axes[1].axvspan(600, 800, color = 'red', alpha = 0.2)

plt.tight_layout()
plt.show()
```
